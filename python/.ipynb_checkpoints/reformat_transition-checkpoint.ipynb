```python
fp_tables_read = os.path.join(sr.dir_data, "transition_info", "SEAD simulation input codes.xlsx")
```

```python
dict_tables = pd.read_excel(fp_tables_read, sheet_name = None)
```

```python
# function for filling in missing labels
def fill_nms(df_in: pd.DataFrame, field_fill: str, missing_flag = "nan", dummy_label = "dummy_label"):
    
    nms = np.array(df_nodes[field_fill]).astype(str)
    w = np.where(nms == missing_flag)[0]
    
    if len(w) > 0:
        fills = [f"{dummy_label}_{x}" for x in range(len(w))]
        np.put(nms, w, fills)
    
    df_in[field_fill] = nms
    
    return 0

def fill_vals(df_in: pd.DataFrame, dict_fills: dict):
    for k in dict_fills.keys():
        df_in[k].fillna(dict_fills[k], inplace = True)

    
# set some parameters
key_nodes = "Nodes"
field_node = "Node"
field_nodelab = "Label"
field_time = "Time"
field_val = "Outcome value at final time"
# some output fields
field_prepend_out = "transition_expression_scenario"
field_scenario_id = "scenario_id"
field_scenario_name = "scenario_name"
# tables to ignore: baseline is repeated in baseline blue stealth
tables_ignore = ["Baseline"]
# replace missing values in the node table with these
dict_repls_na = {field_time: 1, field_val: 0}


# initialize the node table, then fill in names and values
df_nodes = dict_tables[key_nodes].copy()
fill_nms(df_nodes, field_nodelab)
fill_vals(df_nodes, dict_repls_na)


df_transition = df_nodes[[field_node]].copy()

init_dft_q = True

dict_fields_edge = {"To": "i", "From": "j"}
fields_edge = list(dict_fields_edge.keys())

ind_scen_base = 0

scens_inds = []
scens_names = []

for k in [x for x in dict_tables.keys() if (x not in [key_nodes] + tables_ignore)]:
    
    df_transition_cur = dict_tables[k].copy()
    
    prob_scens_0 = [x for x in df_transition_cur.columns if (x not in fields_edge)]
    prob_scens = [f"{k} - {x}" for x in prob_scens_0]
    inds = np.arange(ind_scen_base, ind_scen_base + len(prob_scens)).astype(int)
    
    scens_inds += list(inds)
    scens_names += prob_scens
    
    flds_out = [f"{field_prepend_out}_{x}" for x in inds]
    dict_rnm = dict(zip(prob_scens_0, flds_out))
    
    df_transition_cur.rename(columns = dict_rnm, inplace = True)
    
    if init_dft_q:
        df_transition = df_transition_cur
        init_dft_q = False
    else:
        df_transition = pd.merge(df_transition, df_transition_cur, how = "outer")
    
    ind_scen_base += len(inds)
    print(ind_scen_base)
    
df_transition = df_transition.rename(columns = dict_fields_edge).sort_values(by = list(dict_fields_edge.values())).reset_index(drop = True)
df_transition.fillna(0, inplace = True)
fields_trans_exp = [x for x in df_transition.columns if (x not in dict_fields_edge.values())]

# attribute for scenarios
df_attr_scen = pd.DataFrame({field_scenario_id: scens_inds, field_scenario_name: scens_names})
##  build state attribute
all_scen = list(df_attr_scen[field_scenario_id].unique())
all_scen.sort()
# rename the nodes dataframe
df_nodes.rename(columns = {field_node: "state", field_nodelab: "name", field_time: "time", field_val: "outcome_value"}, inplace = True)

init_q = True
for scen in all_scen:
    
    df_nodes_cur = df_nodes.copy()
    df_nodes_cur = df_nodes_cur.rename(columns = {"time": (f"time_{scen}"), "outcome_value": (f"outcome_value_{scen}")})
    if init_q:
        df_attr_state = df_nodes_cur
        init_q = False
    else:
        df_attr_state = pd.merge(df_attr_state, df_nodes_cur)

df_attr_state = df_attr_state.sort_values(by = ["state"]).reset_index(drop = True)


# exports
df_transition.to_csv(sr.fp_csv_transition, index = None, encoding = "UTF-8")
df_attr_scen.to_csv(sr.fp_csv_attribute_scenario, index = None, encoding = "UTF-8")
df_attr_state.to_csv(sr.fp_csv_attribute_state, index = None, encoding = "UTF-8")



```
